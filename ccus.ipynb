```python
import pandas as pd

# Replace 'your_excel_file.xlsx' with the path to your Excel file
df = pd.read_excel('Downloads\IEA CCUS Projects Database 2024.xlsx', sheet_name = 'Sheet1')

# Now you can work with the DataFrame 'df'
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
data = df['Estimated capacity by IEA (Mt CO2/yr)']

# Plot histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Capacity')
plt.ylabel('Frequency')
plt.title('Capacity (Mt CO2/yr)')

# Show plot
plt.show()
```

```python
column_name = 'Estimated capacity by IEA (Mt CO2/yr)'
mean_column = 'mean_' + column_name

# Convert the column to numeric (if possible)
df[column_name] = pd.to_numeric(df[column_name], errors='coerce')

# Calculate the mean of the non-missing values in the column
mean_value = df[column_name].mean()

# Replace NaN values with the mean
df[mean_column] = df[column_name].fillna(mean_value)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the necessary columns

# Set the style of seaborn
sns.set(style="whitegrid")

# Create scatter plot
sns.scatterplot(data=df, x='Announcement', y='mean_Estimated capacity by IEA (Mt CO2/yr)', hue='Project Status')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Add title and labels
plt.title('Scatter Plot of Announced Capacity vs Announcement with Project Status')
plt.xlabel('Announcement')
plt.ylabel('Announced Capacity (Mt CO2/yr)')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the necessary columns

# Set the style of seaborn
sns.set(style="whitegrid")

# Create scatter plot
sns.scatterplot(data=df, x='Announcement', y='mean_Estimated capacity by IEA (Mt CO2/yr)', hue='Sector')

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Add title and labels
plt.title('Scatter Plot of Announced Capacity vs Announcement with Project Status')
plt.xlabel('Announcement')
plt.ylabel('Announced Capacity (Mt CO2/yr)')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame

# Create a pivot table with 'Sector' as index, 'Announcement' year as columns, and 'mean_Estimated capacity by IEA (Mt CO2/yr)' as values
pivot_df = df.pivot_table(index='Announcement', columns='Sector', values='mean_Estimated capacity by IEA (Mt CO2/yr)', aggfunc='sum')

# Plot the stacked column chart
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add labels and title
plt.title('Stacked Column Chart of Estimated Capacity by IEA')
plt.xlabel('Year')
plt.ylabel('Mean Estimated Capacity (Mt CO2/yr)')
plt.xticks(rotation=45)

# Show plot
plt.legend(title='Announcement Year')
plt.show()
```

```python
import pandas as pd

# Assuming df is your DataFrame

# Group by 'Sector' and calculate the sum of 'mean_Estimated capacity by IEA (Mt CO2/yr)' for each sector
sum_capacity_by_sector = df.groupby('Sector')['mean_Estimated capacity by IEA (Mt CO2/yr)'].sum().reset_index()

# Create a new DataFrame with the results
new_df = pd.DataFrame(sum_capacity_by_sector)

# Display the new DataFrame
new_df.head(15)
```

```python
import matplotlib.pyplot as plt

# Assuming new_df is your DataFrame containing the sum of capacity by sector

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(new_df['mean_Estimated capacity by IEA (Mt CO2/yr)'], labels=new_df['Sector'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Estimated Capacity by Sector')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show plot
plt.show()
```
